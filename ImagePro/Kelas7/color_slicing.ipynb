```python
import cv2 # type: ignore
from cv2 import imread, imshow, waitKey
import matplotlib.pyplot as plt 
import numpy as np
import math
```

```python
def display_image(image, title, size = 5):
    width, height = image.shape[0], image.shape[1]
    aspect_ratio = width/height
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    #plt.imshow(image)
    plt.title(title)
    plt.show()

# Load the image
image = cv2.imread('../Kelas2/flower.jpg')

# Convert the image to the HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the lower and upper bounds for the green color range
lower_bound = np.array([35, 100, 50])   # Lower bound for green
upper_bound = np.array([85, 255, 255])  # Upper bound for green

# Create a mask using the bounds
mask = cv2.inRange(hsv_image, lower_bound, upper_bound)

# Apply the mask to the image to extract the desired color
sliced_image = cv2.bitwise_and(image, image, mask=mask)

# Display the original image and the color-sliced image
display_image(image, 'Original Image')
display_image(sliced_image, 'Color Sliced Image')
display_image(hsv_image, 'HSV Image')



# Split the HSV image into its H, S, and V components
h, s, v = cv2.split(hsv_image)

# Display the individual channels
display_image(h, 'Hue Channel')
display_image(s, 'Saturation Channel')
display_image(v, 'Value Channel')

```
